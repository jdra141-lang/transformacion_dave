Un **boxplot** (o diagrama de caja y bigotes) es una herramienta gráfica que permite visualizar la distribución de un conjunto de datos de manera eficiente. Sus principales beneficios son:

1. **Resumir la distribución de los datos** → Muestra la mediana, cuartiles y valores extremos de un conjunto de datos de forma compacta.
2. **Detectar outliers fácilmente** → Los valores atípicos quedan representados fuera de los "bigotes", facilitando su identificación.
3. **Comparar distribuciones** → Útil para analizar cómo varían los datos entre diferentes categorías o grupos.
4. **Mostrar asimetría y dispersión** → Permite observar si los datos están sesgados y qué tan dispersos están.
5. **Resistencia a valores extremos** → No se ve tan afectado por outliers como otras visualizaciones, como histogramas o medias.

### 5. Prueba de Hipotesis.

### 5.1. Calificacion de los Usuarios para las Plataformas Xbox One y PC.


Hipótesis Nula (H0): Las Califaciones de los usuarios para las plataformas Xbox One y PC son las mismas.

Hipótesis Alternativa (H1): Las Calificaciones de los usuarios para las plataformas Xbox One y PC son diferentes.

<div class="alert alert-block alert-success"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

Has definido correctamente las hipótesis.

</div>

```python

alpha = 0.5
df_xbox_one = df_games[df_games['platform'] == 'XOne']['user_score']
df_pc = df_games[df_games['platform'] == 'PC']['user_score']
results = st.ttest_ind(df_xbox_one, df_pc)
print('valor p:', results.pvalue)
if results.pvalue < alpha:
    print('Rechazamos la hipótesis Nula')
else:
    print('No podemos Rechazar la hipótesis Nula')
```

<div class="alert alert-block alert-warning"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

Al realizar una prueba de hipótesis, es muy importante verificar los supuestos. El t-test que usaste funciona bien cuando se cumplen ciertos supuestos, como que las varianzas de ambos grupos (sábados lluviosos y no lluviosos) sean similares. Puedes realizar la prueba de levene para comprobar que los dos grupos tienen la misma varianza y, de esta forma, podrás saber el valor del parámetro equal_var que se debe utilizar en la función stats.ttest_ind.

### 5.2. Calificación de los Usuarios para los Generos Acción y Deportes.

Hipótesis Nula (H0): Las Califaciones de los usuarios para los generos Acción y Deportes son diferentes.

Hipótesis Alternativa (H1): Las Calificaciones de los usuarios para los Generos Acción y Deportes son iguales.

```python

alpha = 0.5
df_accion = df_games[df_games['genre'] == 'Action']['user_score']
df_deportes = df_games[df_games['genre'] == 'Sports']['user_score']
results = st.ttest_ind(df_accion, df_deportes)
print('valor p:', results.pvalue)
if results.pvalue < alpha:
    print('Rechazamos la hipótesis Nula')
else:
    print('No podemos Rechazar la hipotesis Nula')
```

Tome como base los enunciados de la instrucciones para tomarlos como mis hipótesis nulas, las hipótesis alternativas es lo contrario a la nula. 

Se utilizó st.ttest_ind porque las muestras de puntuaciones de usuario para los juegos de Xbox One y PC, asi como para los generos Acción y Deportes son independientes. Esto significa que los datos de un grupo no están relacionados de ninguna manera con los datos del otro grupo. Por ejemplo, los usuarios de Xbox One son diferentes a los usuarios de PC, y sus puntuaciones no están emparejadas individualmente.

El análisis exhaustivo de los datos de ventas de videojuegos ha revelado tendencias y dinámicas importantes en el mercado, tanto a nivel de plataforma como de género, así como la interacción entre juego y plataforma.

### Hallazgos Importantes:

1.  **Dominio Histórico de Plataformas Antiguas**: Plataformas como PS2, X360 y PS3 demostraron un volumen de ventas totales históricamente elevado, lo que refleja su larga vida útil y la vasta cantidad de títulos lanzados. Sin embargo, algunas plataformas más antiguas pero con menos títulos (ej. GB, NES) mostraron ventas promedio por juego excepcionalmente altas, lo que sugiere el impacto de juegos icónicos y de gran éxito individual.

2.  **Variación en Ventas de Juegos Multiplataforma**: La popularidad de un mismo juego varía significativamente entre diferentes plataformas. Ejemplos como `Grand Theft Auto V` muestran cómo el momento del lanzamiento y la base de usuarios de una consola pueden influir drásticamente en el rendimiento de ventas, con picos en plataformas como PS3 y X360 en su momento.

3.  **Rentabilidad por Género**: Los géneros **Action**, **Sports** y **Shooter** lideran en ventas totales, lo que indica su amplia aceptación y el gran volumen de lanzamientos. En términos de rentabilidad por juego (ventas promedio), los géneros **Platform**, **Shooter** y **Role-Playing** se destacan, sugiriendo que, aunque no siempre tienen la mayor cantidad de títulos, los juegos individuales en estas categorías suelen ser muy exitosos. Los géneros **Adventure** y **Strategy** mostraron ser menos rentables en ambos aspectos.

Esto sugiere que las comunidades de jugadores en estas plataformas pueden tener preferencias o expectativas diferentes, lo que influye en cómo valoran los juegos.

### Implicaciones Generales:

*   **Longevidad vs. Éxito Individual**: El éxito total de una plataforma a menudo se construye sobre una larga historia y un gran catálogo, mientras que el éxito promedio por juego puede estar más relacionado con la calidad intrínseca o el impacto cultural de títulos específicos.
*   **Importancia del Contexto de Plataforma**: Para juegos multiplataforma, es crucial entender la audiencia y las tendencias de cada plataforma para optimizar las estrategias de lanzamiento y marketing. No todos los éxitos se replican de la misma manera en todas partes.
*   **Segmentación del Mercado por Género**: Si bien algunos géneros tienen un atractivo masivo, otros, aunque de nicho, pueden ser muy rentables si los títulos son de alta calidad. La comprensión de estas dinámicas es vital para el desarrollo y la inversión en juegos.
*   **Experiencia del Usuario Específica de la Plataforma**: Las diferencias en las puntuaciones de usuario entre plataformas destacan la importancia de adaptar la experiencia del juego y su lanzamiento a las expectativas y características de la base de jugadores de cada sistema.
